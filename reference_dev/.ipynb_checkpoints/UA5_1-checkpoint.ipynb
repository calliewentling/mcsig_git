```python
#Exemplo 16

import os
from osgeo import ogr, osr

path = os.path.join('/home', 'jovyan', 'geonames')
os.chdir(path)

wgs = osr.SpatialReference()
wgs.ImportFromEPSG(4326)

driver = ogr.GetDriverByName('ESRI Shapefile')
shp = driver.CreateDataSource('geonames.shp')
layer = shp.CreateLayer('geonames', srs=wgs, geom_type=ogr.wkbPoint)

fields = [{'name': 'geonameid', 'type': ogr.OFTInteger},
         {'name': 'name', 'type': ogr.OFTString, 'width': 200},
         {'name': 'latitude', 'type': ogr.OFTReal},
         {'name': 'longitude', 'type': ogr.OFTReal},
         {'name': 'population', 'type': ogr.OFTInteger},
         {'name': 'dem', 'type': ogr.OFTInteger}]

for i in fields:
    defn = ogr.FieldDefn(i['name'], i['type'])
    if 'width' in i:
        defn.SetWidth(i['width'])
    layer.CreateField(defn)

shp.Destroy()
```

```python
#Exemplo 47
import os
from osgeo import ogr, osr
from shapely import geometry, wkt

def addItems(targetList, lyr):
    feature = lyr.GetNextFeature()
    while feature:
        targetList.append(feature)
        feature = lyr.GetNextFeature()
    return targetList

path = os.path.join('/home', 'jovyan', 'natural_earth')
os.chdir(path)

wgs = osr.SpatialReference()
wgs.ImportFromEPSG(4326)
etrs = osr.SpatialReference()
etrs.ImportFromEPSG(3763)
reproj = osr.CoordinateTransformation(wgs,etrs)

shp = ogr.Open('countries.shp')
layer = shp.GetLayer(0)
layer.SetAttributeFilter("SOVEREIGNT =  'Portugal'")
pt = layer.GetNextFeature()
ptGeom = pt.GetGeometryRef()
minX,maxX,minY,maxY = ptGeom.GetEnvelope()
ptGeom.Transform(reproj)
shp.Destroy()

shp = ogr.Open('rivers.shp')
layer = shp.GetLayer(0)
layer.SetSpatialFilterRect(minX,minY,maxX,maxY)
all_rivers = addItems([], layer)
shp.Destroy()

shp = ogr.Open('rivers_europe.shp')
layer = shp.GetLayer(0)
layer.SetSpatialFilterRect(minX,minY,maxX,maxY)
all_rivers = addItems(all_rivers, layer)
shp.Destroy()

driver = ogr.GetDriverByName('ESRI Shapefile')
shpFinal = driver.CreateDataSource('rivers_pt_buffer.shp')
layerFinal = shpFinal.CreateLayer('rivers_pt_buffer', etrs)
field = ogr.FieldDefn('name', ogr.OFTString)
field.SetWidth(50)
layerFinal.CreateField(field)

memDriver = ogr.GetDriverByName('MEMORY')
memSource = memDriver.CreateDataSource('memory')
memLayer = memSource.CreateLayer('memory', etrs)

ptFeat = ogr.Feature(memLayer.GetLayerDefn())
ptFeat.SetGeometry(ptGeom)
memLayer.CreateFeature(ptFeat)

for i in all_rivers:
    geom = i.GetGeometryRef()
    geom.Transform(reproj)
    geomWkt = geom.ExportToWkt()
    shapelyGeom = wkt.loads(geomWkt)
    buffer = shapelyGeom.buffer(1500)
    bufWkt = wkt.dumps(buffer)
    bufGeom = ogr.CreateGeometryFromWkt(bufWkt, etrs)
    
    newFeat = ogr.Feature(layerFinal.GetLayerDefn())
    newFeat.SetField('name', i.GetFieldAsString('name'))
    newFeat.SetGeometry(bufGeom)
    layerFinal.CreateFeature(newFeat)
    newFeat.Destroy()
    
shpClip = driver.CreateDataSource('rivers_pt_clip.shp')
layerClip = shpClip.CreateLayer('rivers_pt_clip.shp')

layerFinal.Clip(memLayer, layerClip)

memSource.Destroy()
shpClip.Destroy()
shpFinal.Destroy()
```

```python
#Exemplo 51

import os
import psycopg2
from osgeo import ogr, osr

path = os.path.join('/home', 'jovyan', 'outputs')
#os.mkdir(path)
os.chdir(path)

import psycopg2

conn = psycopg2.connect('host=postgres ' + 'dbname=pypg user=postgres ' + 'password=postgres')
cur = conn.cursor()

cur.execute("SELECT distrito, " + 
           "st_astext(st_union(geom)) as geom " + 
           " FROM caop GROUP BY distrito")
records = cur.fetchall()

etrs = osr.SpatialReference()
etrs.ImportFromEPSG(3763)

driver = ogr.GetDriverByName('ESRI Shapefile')
shp = driver.CreateDataSource('distritos.shp')
layer = shp.CreateLayer('distritos', etrs)  #'NoneType' object has no attribute 'CreateLayer'

field = ogr.FieldDefn('concelho', ogr.OFTString)
field.SetWidth(100)
layer.CreateField(field)

for rec in records:
    geom = ogr.CreateGeometryFromWkt(rec[1])
    feature = ogr.Feature(layer.GetLayerDefn())
    feature.SetField('concelho', rec[0])
    feature.SetGeometry(geom)
    layer.CreateFeature(feature)
    feature.Destroy()

shp.Destroy()
cur.close()
conn.close()
    
```

```python
#Exemplo 56

import os
from osgeo import gdal

path = os.path.join('/home', 'jovyan', 'srtm')
os.chdir(path)

raster = gdal.Open('srtm_35_05.tif')
band = raster.GetRasterBand(1)
geoTransf = raster.GetGeoTransform()
invTransf = gdal.InvGeoTransform(geoTransf)

path = os.path.join('/home', 'jovyan', 'geonames')
os.chdir(path)

shp = ogr.Open('geonames.shp', 1)
layer = shp.GetLayer(0)

feature = layer.GetNextFeature()
while feature:
    geom = feature.GetGeometryRef()
    xPx, yPx = gdal.ApplyGeoTransform(invTransf, geom.GetX(), geom.GetY())
    value = band.ReadAsArray(int(xPx),int(yPx),1,1)
    if value is not None:
        feature.SetField('elevation', int(value[0][0]))
        layer.SetFeature(feature)
    feature = layer.GetNextFeature()

band = None
raster = None
shp.Destroy()
```

```python
#Exemplo 57

import os
from osgeo import gdal, osr

path = os.path.join('/home', 'jovyan', 'srtm')
os.chdir(path)

raster = gdal.Open('srtm_35_05.tif')
rastSrsWkt = raster.GetProjection()

driver = gdal.GetDriverByName('GTiff')
driver.Register()

outSrs = osr.SpatialReference()
outSrs.ImportFromEPSG(3763)
outSrsWkt = outSrs.ExportToWkt()

reproj = gdal.AutoCreateWarpedVRT(raster, rastSrsWkt, outSrsWkt)
gdal.ReprojectImage(raster, reproj, rastSrsWkt, outSrsWkt)

rastReproj = driver.CreateCopy('srtm_35_05_etrs.tif', reproj)

raster = None
rastReproj = None
```

```python
#Exemplo 58

import os
from osgeo import gdal, osr

path = os.path.join('/home', 'jovyan', 'srtm')
os.chdir(path)

gdal.AllRegister()


#Abrir o primeiro ficheiro e extrair o número colunas e linhas,
# o minx, miny, maxx, maxy, resolução em X e Y

rast1 = gdal.Open('srtm_35_05.tif')
band1 = rast1.GetRasterBand(1)
rows1 = rast1.RasterYSize
cols1 = rast1.RasterXSize
minx1, pxwidth1, xrot1, maxy1, yrot1, pxheigth1 = rast1.GetGeoTransform()
maxx1 = minx1 + (cols1*pxwidth1)
miny1 = maxy1 + (rows1*pxheigth1)

#Abrir o segundo ficheiro e extrair o número colunas e linhas,
# o minx, miny, maxx, maxy, resolução em X e Y
rast2 = gdal.Open('srtm_35_04.tif')
band2 = rast2.GetRasterBand(1)
rows2 = rast2.RasterYSize
cols2 = rast2.RasterXSize
minx2, pxwidth2, xrot2, maxy2, yrot2, pxheigth2 = rast2.GetGeoTransform()
maxx2 = minx2 + (cols2*pxwidth2)
miny2 = maxy2 + (rows2*pxheigth2)

#Caclular as coordenadas mínimas e máximas da imagem final
# e número de linhas e colunas
finalmaxx = max(maxx1, maxx2)
finalmaxy = max(maxy1, maxy2)
finalminx = min(minx1, minx2)
finalminy = min(miny1, miny2)
finalcols = int((finalmaxx-finalminx)/pxwidth1)
finalrows = int((finalmaxy-finalminy)/pxheigth1)

#Calcular os offsets em X e Y para cada imagem
xoff1 = int((minx1-finalminx)/pxwidth1)
xoff2 = int((minx2-finalminx)/pxwidth1)
yoff1 = abs(int((maxy1-finalmaxy)/pxheigth1))
yoff2 = int((maxy2-finalmaxy)/pxheigth1)

#obter driver, criar ficheiro com as novas dimensões e resoluções
# e definir o NoDataValue

driver = rast1.GetDriver()
outrast = driver.Create('srtm_merge_python.tif', finalcols, finalrows, 1, band1.DataType) #Need a second input as pointer?
bandout = outrast.GetRasterBand(1) #'NoneType' object has no attribute 'GetRasterBand'
bandout.SetNoDataValue(-32768)

#Ler os daddos do primeiro ficheiro copiá-los para a imagem final
data1 = band1.ReadAsArray(0,0,cols1,rows1)
bandout.WriteArray(data1, xoff1, yoff1)

#Ler os dados do segundo ficheiro copiá-los para a imagem final
data2 = band2.ReadAsArray(0,0,cols2,rows2)
bandout.WriteArray(data2, xoff2, yoff2)

#Guardar dados e limpar memória, forçar cálculo de estatísticas
# Definir as propriedades do GeoTransform
# Definir projeção
# Construir overviews

bandout.FlushCache()
stats = bandout.GetStatistics(0, 1)
geotransf = (finalminx, pxwidth1, 0, finalmaxy, 0, pxheigth1)
outrast.SetGeoTransform(geoTransf)
outrast.SetProjection(rast1.GetProjection())
outrast.BuildOverviews(overviewlist=[2,4,8,16,32,64,128])

rast1 = None
rast2 = None
outrast = None
```
